```python
import matplotlib.pyplot as plt

# Sample data for energy consumption over 6 months (in MWh)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
energy_consumption = [1200, 1300, 1100, 1500, 1400, 1600]

# Create a line plot
plt.plot(months, energy_consumption, marker='s', color='r', linestyle='solid')

# Add titles and labels
plt.title('Energy Consumption Over 6 Months')
plt.xlabel('Month')
plt.ylabel('Energy Consumption (MWh)')
plt.show()
```

```python
# Sample data for energy consumption by different renewable sources
energy_sources = ['Solar', 'Wind', 'Hydropower', 'Biomass']
energy_values = [1200, 3400, 2900, 2500]

# Create a bar chart
plt.bar(energy_sources, energy_values, color='green',)

# Add titles and labels
plt.title('Energy Consumption by Renewable Energy Source')
plt.xlabel('Energy Source')
plt.ylabel('Energy Consumption (MWh)')
plt.show()
```

```python
# Pie chart to show energy consumption distribution
plt.pie(energy_values, labels=energy_sources, autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen', 'coral'])

# Add a title
plt.title('Energy Consumption Share by Source')
plt.show()
```

```python
cars = ['AUDI', 'BMW', 'FORD', 'TESLA', 'JAGUAR', 'MERCEDES']
data = [23, 17, 35, 29, 12, 41]

fig = plt.figure(figsize=(10, 7))
plt.pie(data, labels = cars, explode = (0, 0, 0, 0, 0, 0), shadow=True)

plt.show()
```

```python
import numpy as np
x = np.arange(0,10,0.1)
y = np.sin(x)

plt.figsize=(10, 10)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,10,0.1)
y = np.sin(x)
z = np.cos(x)
a = np.tan(x)
b = 1/np.tan(x)

plt.figure(figsize=(10, 7))
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('Sine Function')

plt.subplot(2, 2, 2)
plt.plot(x, z)
plt.title('Cosine Function')

plt.subplot(2, 2, 3)
plt.plot(x, a)
plt.title('Tangent Function')

plt.subplot(2, 2, 4)
plt.plot(x, b)
plt.title('Cotangent Function')

plt.show()
```
